```python
import random
import numpy as np
import math
from scipy.stats import bernoulli 
```

```python
#экспоненциальное с параметром 1
def g(x):
    return(math.exp(-x))
```

```python
#плотность для 2-го шага
def f(x):
    return(math.exp(-1/2*x**2-1/2+x))
```

```python
#генерируем случайную величину с экспоненциальным распределением
def g_obr(x):
    return(-math.log(x))
```

```python
#шаг 1: генерируем случайную величину Y с плотностью g(x)
def generation1():
    x = random.uniform(0, 1)
    y = g_obr(x)
    return(y)
```

```python
generation1()
```

```python
#шаг 2: разыгрываем бернуллиевскую величину с вероятностью f(y)
def generation2():
    x = generation1()
    p = f(x) 
    t = np.random.binomial(1,p)
    if (t == 1):
        return(x)
    else: 
        return(generation2())
```

```python
generation2()
```

```python
a = np.zeros(1000)
for i in range(0,999):
    a[i] = generation2()
```

```python
#получили массив из модулей нормальных величин
a
```
